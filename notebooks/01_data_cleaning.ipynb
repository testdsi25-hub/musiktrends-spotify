```python
import pandas as pd
import glob
```

```python
# in den Ordner navigieren
files = glob.glob("../data/raw/spotify-weekly/*.csv")
```

```python
dfs = []

for f in files:
    df = pd.read_csv(f)
    df["source_file"] = f
    dfs.append(df)
```

```python
if dfs: 
    charts = pd.concat(dfs, ignore_index=True)
    print(f"Erfolgreich {len(files)} Dateien geladen.")
    display(charts.head())
else:
    print("Keine Dateien gefunden. Prüfe den Pfad.")
```
